```python
import tkinter as tk
from tkinter import messagebox, simpledialog
from datetime import datetime
import os
import csv

students = []


filename = "attendance.csv"

def get_current_date():
    return datetime.now().strftime("%Y-%m-%d")

# Function to check if attendance is already recorded
def is_attendance_recorded(student_name, date):
    try:
        if os.path.exists(filename):
            with open(filename, "r", newline='') as file:
                reader = csv.reader(file)
                for row in reader:
                    # Check if the student name and date already exist
                    if student_name == row[1] and date == row[2]:
                        return True
    except Exception as e:
        messagebox.showerror("Error", f"Could not check attendance records: {e}")
    return False

# Function to add a new attendance record
def add_attendance(student_id, student_name, date):
    # First, check if the record already exists
    if is_attendance_recorded(student_name, date):
        messagebox.showinfo("Info", f"Attendance already recorded for {student_name} on {date}.")
        return

    # If the attendance hasn't been recorded, add it to the file
    try:
        # Check if the file exists, and if not, create it with a header
        file_exists = os.path.exists(filename)
        
        with open(filename, "a", newline='') as file:  # "a" mode opens the file for appending
            writer = csv.writer(file)
            
            # Write the header if the file does not exist yet
            if not file_exists:
                writer.writerow(["Student ID", "Student Name", "Date"])

            # Add the attendance record
            writer.writerow([student_id, student_name, date])
            messagebox.showinfo("Success", f"Attendance recorded for {student_name} on {date}.")
    except Exception as e:
        messagebox.showerror("Error", f"Could not write attendance record: {e}")

def refresh_student_list():
    # Clear existing frames
    for widget in frame.winfo_children():
        widget.destroy()

    # Display each student with a "Mark Attendance" button
    for student in students:
        student_frame = tk.Frame(frame)
        student_frame.pack(fill="x", pady=5)

        # Student name label
        name_label = tk.Label(student_frame, text=student["name"], font=("Arial", 12), width=20, anchor="w")
        name_label.pack(side="left")

        # Mark attendance button
        mark_button = tk.Button(student_frame, text="Mark Attendance", command=lambda s=student: mark_attendance(s["id"], s["name"]))
        mark_button.pack(side="right")

# Function to add a new student
def add_new_student():
    try:
        new_id = simpledialog.askinteger("Input", "Enter Student ID")
        new_name = simpledialog.askstring("Input", "Enter Student Name")

        if new_id and new_name:
            # Check if the student ID already exists
            if any(student['id'] == new_id for student in students):
                messagebox.showerror("Error", "Student ID already exists!")
            else:
                # Add the new student to the list
                students.append({"id": new_id, "name": new_name})
                refresh_student_list()  # Refresh the list to show the new student
        else:
            messagebox.showwarning("Input Error", "Please provide valid Student ID and Name.")
    except Exception as e:
        messagebox.showerror("Error", f"Could not add new student: {e}")

# Tkinter application
root = tk.Tk()
root.title("Student Attendance System")
root.geometry("400x400")

# Display today's date
date_label = tk.Label(root, text=f"Date: {get_current_date()}", font=("Arial", 14))
date_label.pack(pady=10)

# Frame for student list
frame = tk.Frame(root)
frame.pack(pady=10)

# Button to add new students
add_student_button = tk.Button(root, text="Add New Student", command=add_new_student)
add_student_button.pack(pady=10)

# Initial display of student list
refresh_student_list()

# Start the Tkinter main loop
root.mainloop()

# Example usage


# Add a new attendance record
```

```python
import tkinter as tk
from tkinter import messagebox, simpledialog
from datetime import datetime
import os
import csv
from pymongo import MongoClient






# Sample student data
students = [
    {"id": 1, "name": "John Doe"},
    {"id": 2, "name": "Jane Smith"},
    {"id": 3, "name": "Alice Johnson"},
    {"id": 4, "name": "Bob Brown"}
]

# In-memory attendance records
attendance_records = []

# Function to get today's date
def get_current_date():
    return datetime.now().strftime("%Y-%m-%d")

# Function to check if attendance is already recorded
def is_attendance_recorded(student_name, date):
    for record in attendance_records:
        if record['name'] == student_name and record['date'] == date:
            return True
    return False

# Function to mark attendance
def mark_attendance(student_id, student_name):
    date = get_current_date()
    
    # Check if attendance is already marked
    if is_attendance_recorded(student_name, date):
        messagebox.showinfo("Attendance", f"Attendance for {student_name} already marked.")
        return

    # Add attendance record to the list
    attendance_records.append({"id": student_id, "name": student_name, "date": date})
    messagebox.showinfo("Attendance", f"Attendance marked for {student_name}.")
    print(f"Attendance for {student_name} recorded in memory list.")

# Function to refresh the student list
def refresh_student_list():
    for widget in frame.winfo_children():
        widget.destroy()

    for student in students:
        student_frame = tk.Frame(frame,bg='#333333')
        student_frame.pack(fill="x", pady=5)

        name_label = tk.Label(student_frame, text=student["name"], font=("Arial", 12), width=20, anchor="w",fg='white',bg='#333333')
        name_label.pack(side="left")

        mark_button = tk.Button(student_frame, text="Mark Attendance", command=lambda s=student: mark_attendance(s["id"], s["name"]),fg='white',bg='#333333')
        mark_button.pack(side="right")

# Function to add a new student
def add_new_student():
    try:
        new_id = simpledialog.askinteger("Input", "Enter Student ID")
        new_name = simpledialog.askstring("Input", "Enter Student Name")

        if new_id and new_name:
            if any(student['id'] == new_id for student in students):
                messagebox.showerror("Error", "Student ID already exists!",fg='white',bg='#333333')
            else:
                students.append({"id": new_id, "name": new_name})
                refresh_student_list()
        else:
            messagebox.showwarning("Input Error", "Please provide valid Student ID and Name.",fg='white',bg='#333333')
    except Exception as e:
        messagebox.showerror("Error", f"Could not add new student: {e}")

# Tkinter application

root = tk.Tk()
root.title("Student Attendance System")
root.geometry("400x400")
root.configure(bg='#333333')

date_label = tk.Label(root, text=f"Date: {get_current_date()}", font=("Arial", 14),fg='white',bg='#333333')
date_label.pack(pady=19)

frame = tk.Frame(root,bg="#333333")
frame.pack(pady=10)

add_student_button = tk.Button(root, text="Add New Student", command=add_new_student,fg='white',bg='#333333')
add_student_button.pack(pady=10)

refresh_student_list()

# MongoDB initialization (place before the Tkinter mainloop or where appropriate)
uri = "mongodb+srv://mangeshsambare1:mangeshsambare123@cluster0.cw3cocl.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
client = MongoClient(uri)
database = client['Attendance_sheet']
coll = database['Daily_Attendance']
# Start the Tkinter main loop
root.mainloop()

# After the Tkinter window closes, save the attendance records to MongoDB
if attendance_records:
    coll.insert_many(attendance_records)
    print("Attendance records saved to MongoDB.")
```
